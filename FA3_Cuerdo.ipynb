```python
import numpy as np
import scipy.stats as stats
import pandas as pd
from IPython.display import display

data = [88, 45, 53, 86, 33, 86, 85, 30, 89, 53, 41, 96, 56, 38, 62,
        71, 51,86, 68, 29, 28, 47, 33, 37, 25, 36, 33, 94, 73, 46,
        42, 34, 79, 72, 88, 99, 82, 62, 57, 42, 28, 55, 67, 62, 60,
        96, 61, 57, 75, 93, 34, 75, 53, 32, 28, 73, 51, 69, 91, 35 
       ]

data_np = pd.DataFrame(data, columns = ['Score'])

valid = len(data_np)
mean = data_np['Score'].mean()
std_dv = data_np['Score'].std(ddof=1)
var = data_np['Score'].var(ddof=1)
skewness = stats.skew(data_np['Score'])
kurtosis = stats.kurtosis(data_np['Score'])
min_val = data_np['Score'].min()
max_val = data_np['Score'].max()
mode_val = stats.mode(data_np['Score'], keepdims=True)[0][0]
median = data_np['Score'].median()

Q1 = np.percentile(data_np['Score'], 25)
Q2 = np.percentile(data_np['Score'], 50)
Q3 = np.percentile(data_np['Score'], 75)
D9 = np.percentile(data_np['Score'], 90)
P95 = np.percentile(data_np['Score'], 95)

summary = {
    'Valid': valid,
    'Mode': mode_val,
    'Median': median,
    'Mean': mean,
    'Std. deviation': std_dv,
    'Variance': var,
    'Skewness': skewness,
    'Kurtosis': kurtosis, 
    'Minimum': min_val, 
    'Maximum': max_val, 
    'Q1 (25th percentile)': Q1,
    'Q2 (50th percentile)': Q2,
    'Q3 (75th percentile)': Q3,
    'D9 (90th percentile)': D9,
    'P95 (95th percentile)': P95
}

summary_df = pd.DataFrame(summary, index=['Score']).transpose()
summary_df = summary_df.round(3)

display(summary_df)
```
